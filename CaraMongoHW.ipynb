```python
document = collection.find_one()
pprint.pprint(document)
```

```python
# Ex.1 #1
firstaction = collection.find_one({"genres": "Action"})
pprint.pprint(firstaction)
```

```python
# Ex.1 #2
moviesafter2000 = collection.find({"year": {"$gt": 2000}}).limit(5)
for movies in moviesafter2000:
  pprint.pprint(movies)
```

```python
# Ex.1 #3
ratingover85 = collection.find({"imdb.rating": {"$gt": 8.5}}).limit(5)
for movie in ratingover85:
  pprint.pprint(movie)
```

```python
# Ex.1 #4
actionadv = collection.find({"genres": {"$all": ["Action", "Adventure"]}}).limit(5)
for movie in actionadv:
  pprint.pprint(movie)
```

```python
# Ex.2 #1
sortedcom = collection.find({"genres": "Comedy"}).sort("imdb.rating", -1).limit(5)
for movie in sortedcom:
  pprint.pprint(movie)
```

```python
# Ex.2 #2
sorteddram = collection.find({"genres": "Drama"}).sort("year", 1).limit(5)
for movie in sorteddram:
  pprint.pprint(movie)
```

```python
# Ex.3 #1
avggenrerating = collection.aggregate([
    {"$unwind": "$genres"},
     {"$group": {"_id": "$genres", "avg_rating": {"$avg": "$imdb.rating"}}},
      {"$sort": {"avg_rating": -1}},
       {"$limit": 5}
])
for genre in avggenrerating:
  pprint.pprint(genre)
```

```python
# Ex.3 #2
toprateddirectors = collection.aggregate([
    {"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
     {"$sort": {"avg_rating": -1}},
      {"$limit": 5}
])
for director in toprateddirectors:
  pprint.pprint(director)
```

```python
# Ex.3 #3
movperyr = collection.aggregate([
    {"$group": {"_id": "$year", "total_movies": {"$sum": 1}}},
     {"$sort": {"_id": 1}}
])
for year in movperyr:
  pprint.pprint(year)
```

```python
# Ex.4 #1
collection.update_one({"title": "The Godfather"}, {"$set": {"imdb.rating": 9.5}})
```

```python
# Ex.4 #2
collection.update_many({"genres": "Horror", "imdb.rating": {"$exists": False}},
 {"$set":{"imdb.rating": 6.0}})
# updatedExisting = False
```

```python
horrormovies = collection.find({"genres": "Horror", "imdb.rating": {"$exists": False}})
for movie in horrormovies:
    print(movie)
# Checking to see if there are movies in the Horror genre, there are none
```

```python
# Ex.4 #3
collection.delete_many({"year": {"$lt": 1950}})
```

```python
# Ex.5 #1
collection.create_index([("title", "text")])
lovetitle = collection.find({"$text": {"$search": "love"}})
for movie in lovetitle:
    print(movie['title'])
```

```python
# Ex.5 #2
collection.drop_index("title_text")
collection.create_index([("title", "text"), ("plot", "text")])
warmovies = collection.find({"$text": {"$search": "war"}}).sort("imdb.rating", -1).limit(5)
for movie in warmovies:
  print(movie)
```
